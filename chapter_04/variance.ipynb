# 표본의 평균이 모 평균과 다름

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. 5개의 표본 데이터 (우리 동네 5명의 월급이라고 가정)
sample_data = np.array([200, 220, 250, 260, 320])

# 2. 표본평균(x_bar) 계산
x_bar = np.mean(sample_data) # 딱 250이 나옵니다.

# 가상의 진짜 평균(mu)이 270이라고 가정해 봅니다.
mu = 270 

# 3. 임의의 중심점(c)이 200에서 300까지 변할 때, 편차의 제곱합을 계산하는 함수
def sum_of_squared_errors(c, data):
    return np.sum((data - c)**2)

c_values = np.linspace(200, 300, 100)
sse_values = [sum_of_squared_errors(c, sample_data) for c in c_values]

# 4. 시각화
plt.figure(figsize=(10, 6))
plt.plot(c_values, sse_values, color='purple', linewidth=2.5, label='Sum of Squared Errors (편차 제곱합)')

# 표본평균(x_bar) 위치 표시 - 포물선의 최저점!
plt.plot(x_bar, sum_of_squared_errors(x_bar, sample_data), 'ro', markersize=8)
plt.axvline(x_bar, color='red', linestyle='--', label=f'Sample Mean (x_bar = {x_bar}) : MINIMUM')

# 진짜 평균(mu) 위치 표시 - 최저점보다 붕 떠 있음!
plt.plot(mu, sum_of_squared_errors(mu, sample_data), 'bo', markersize=8)
plt.axvline(mu, color='blue', linestyle='-.', label=f'True Mean (mu = {mu}) : HIGHER')

plt.title('Why Sample Variance is Biased (Always Smaller)')
plt.xlabel('Chosen Center Point (c)')
plt.ylabel('Sum of Squared Deviations')
plt.legend()
plt.grid(True)
plt.show()
```
