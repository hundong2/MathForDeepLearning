### 1. 유클리드 거리(Euclidean Distance)란 무엇인가?

**유클리드 거리**는 우리가 일상생활에서 자(Ruler)를 대고 재는 **'가장 짧은 직선 거리'**를 의미합니다. 기원전 300년경 고대 그리스의 수학자 유클리드가 집대성한 기하학의 기초이자, 여러분이 중학교 때 배운 '피타고라스의 정리'를 2차원, 3차원, 그리고 다차원으로 확장한 것입니다.

**수학적 정의:**
두 점 $\mathbf{p} = (p_1, p_2, \dots, p_n)$와 $\mathbf{q} = (q_1, q_2, \dots, q_n)$ 사이의 유클리드 거리 $d$는 다음과 같습니다.
$$d(\mathbf{p}, \mathbf{q}) = \sqrt{\sum_{i=1}^{n} (q_i - p_i)^2}$$

* **본질적 한계:** 유클리드 거리는 모든 변수(축)를 **'동일한 비중과 단위'**로 취급합니다. 데이터가 어떻게 뭉쳐 있든, 어떤 방향으로 퍼져 있든 전혀 신경 쓰지 않고 오로지 물리적인 절대 위치의 차이만 계산합니다.

### 2. 시각적 차이: 동심원 vs 타원 (등고선의 마법)

이 두 거리가 데이터를 바라보는 시각의 차이는 지도의 **등고선(Contour)**으로 상상하면 가장 완벽하게 이해할 수 있습니다. 

데이터의 중심(평균)에서 시작하여 **"거리가 1인 지점, 2인 지점, 3인 지점..."**을 선으로 연결해 본다고 상상해 보십시오.



**상황 가정:** x축은 '키', y축은 '몸무게'입니다. 사람들의 데이터는 우상향하는 길쭉한 **타원형 구름**을 만듭니다. (키가 클수록 몸무게가 많이 나가는 강한 양의 상관관계)

* **유클리드 거리의 눈 (원형 등고선):**
    유클리드 거리는 데이터의 형태를 깡그리 무시합니다. 중심에서 사방으로 똑같이 퍼져나가는 완벽한 **동심원(Circle)** 모양의 등고선을 그립니다. 
    * 결과: 키와 몸무게가 비례해서 같이 커진 사람(정상 패턴)이나, 키는 그대로인데 몸무게만 비정상적으로 늘어난 사람(이상 패턴)이나, 중심에서 직선으로 잰 절대 거리가 같다면 **"둘 다 똑같이 특이하다"**고 판단해 버립니다.
* **[마할라노비스 거리](./Mahalanobis%20Distance.ipynb)의 눈 (타원형 등고선):**
    [마할라노비스 거리](./Mahalanobis%20Distance.ipynb)는 데이터 구름의 형태를 그대로 따라갑니다. 구름의 모양에 맞춘 **타원형(Ellipse)** 등고선을 그립니다.
    * 결과: 데이터가 길게 늘어선 방향(정상 패턴)으로는 등고선 간격이 넓습니다. 즉, 키와 몸무게가 같이 증가하는 것은 '흔한 일'이므로 거리를 가깝게(정상으로) 측정합니다.
    * 반면, 데이터가 좁은 방향(패턴을 벗어난 방향, 예: 키는 190cm인데 몸무게가 40kg)으로는 등고선 간격이 매우 촘촘합니다. 조금만 벗어나도 거리가 기하급수적으로 멀어지게 측정하여 **"패턴이 붕괴된 치명적인 이상치"**로 정확히 잡아냅니다.

### 1. 역사적 배경: "유클리드와 평탄한 우주의 탄생"

**유클리드(Euclid of Alexandria, 기원전 300년경)**는 고대 그리스의 위대한 수학자로, '기하학의 아버지'라 불립니다. 

**왜 기하학을 집대성해야 했는가?**
당시 고대 이집트에서는 나일강이 매년 범람하여 농경지의 경계선이 사라지곤 했습니다. 세금을 걷기 위해 땅의 넓이와 경계를 다시 측량하는 기술(Geo-metry: 땅을 잰다는 뜻)이 절실하게 필요했습니다. 수많은 측량 기술과 파편화된 수학 지식들이 흩어져 있었는데, 유클리드는 이를 모아 역사상 가장 위대한 수학 책인 **《원론(Elements)》**을 집필했습니다.

유클리드의 가장 큰 업적은 **'공리적 방법(Axiomatic Method)'**을 도입한 것입니다. 그는 "두 점을 잇는 직선은 단 하나뿐이다"와 같이 누구도 부정할 수 없는 가장 단순하고 명백한 5개의 기본 가정(공리)을 세우고, 오직 이 공리들만을 바탕으로 논리적인 증명을 통해 피타고라스의 정리 등 수백 개의 복잡한 기하학적 진리들을 도출해 냈습니다.

우리가 일상에서 인지하는 찌그러짐이나 휘어짐이 없는 '평탄한 공간'을 **'유클리드 공간(Euclidean Space)'**이라고 부르며, 이 공간에서 두 점 사이의 '가장 짧은 직선 경로'가 바로 **유클리드 거리**입니다.

### 2. 수학적 정의와 증명: "피타고라스에서 N차원으로"

유클리드 거리는 철저하게 **피타고라스의 정리(Pythagorean Theorem)**에 뿌리를 두고 있습니다. "직각삼각형에서 빗변의 길이의 제곱은 나머지 두 변의 길이의 제곱의 합과 같다"는 원리입니다.



#### 2차원 공간에서의 증명
2차원 평면 좌표계에 점 $A(x_1, y_1)$과 점 $B(x_2, y_2)$가 있다고 가정해 봅시다.
점 A에서 점 B로 가는 가장 짧은 직선거리를 빗변으로 하는 가상의 직각삼각형을 그릴 수 있습니다.
* 가로 변의 길이는 $(x_2 - x_1)$
* 세로 변의 길이는 $(y_2 - y_1)$

피타고라스의 정리에 의해 두 점 사이의 거리 $d$는 다음과 같이 증명됩니다.
$$d^2 = (x_2 - x_1)^2 + (y_2 - y_1)^2$$
$$d = \sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2}$$

#### N차원 공간(다차원 벡터 공간)으로의 확장
통계학과 AI에서 데이터는 2차원(x, y)에 머물지 않습니다. 키, 몸무게, 나이, 소득 등 변수의 개수가 $n$개로 늘어나면 이는 $n$차원 공간의 점(벡터)이 됩니다. 유클리드는 놀랍게도 이 피타고라스의 정리를 단순히 $n$번 더하는 것만으로 다차원 공간의 거리를 정의할 수 있도록 기초를 닦아두었습니다.

점 $\mathbf{p} = (p_1, p_2, \dots, p_n)$와 점 $\mathbf{q} = (q_1, q_2, \dots, q_n)$ 사이의 거리는 다음과 같습니다. AI 분야에서는 이를 **$L_2$ Norm(엘투 노름)**이라는 단축어로도 부릅니다.
$$d(\mathbf{p}, \mathbf{q}) = \sqrt{\sum_{i=1}^{n} (q_i - p_i)^2}$$

### 3. 직관적인 계산 예제

수식을 구체적인 예시로 풀어보겠습니다. 당신이 쇼핑몰 AI를 만들고 있고, 고객 A와 고객 B의 '취향 유사도'를 계산해야 한다고 가정해 봅시다. 
변수는 3개(3차원 공간)이며, 점수는 10점 만점입니다.

* **고객 A (전자제품, 의류, 식품):** $(8, 2, 5)$
* **고객 B (전자제품, 의류, 식품):** $(6, 5, 4)$

두 고객 사이의 유클리드 거리를 계산해 봅시다.
1. 각 차원별 차이를 구합니다: $(8-6, 2-5, 5-4) = (2, -3, 1)$
2. 각 차이를 제곱합니다: $(2^2, (-3)^2, 1^2) = (4, 9, 1)$
3. 제곱한 값들을 모두 더합니다: $4 + 9 + 1 = 14$
4. 마지막으로 루트(제곱근)를 씌웁니다: $\sqrt{14} \approx 3.74$

두 고객의 취향 공간에서의 거리는 약 **3.74**입니다. 이 거리가 짧을수록(0에 가까울수록) AI는 두 고객의 성향이 매우 비슷하다고 판단합니다.

### 4. 실무와 AI에서의 응용: 어디에 쓰이는가?

앞서 배운 [마할라노비스 거리](./Mahalanobis%20Distance.ipynb)가 데이터의 '상관관계'를 볼 때 쓴다면, 유클리드 거리는 **'절대적인 값의 차이'** 그 자체가 중요할 때 쓰이는 가장 기본적이고 연산 속도가 빠른 척도입니다.

* **K-Nearest Neighbors (KNN, K-최근접 이웃 알고리즘):** 새로운 데이터가 들어왔을 때, 기존 데이터들 중 유클리드 거리가 가장 가까운 $K$개의 이웃을 찾아 그 이웃들이 가진 정답으로 다수결 투표를 하는 기초적인 머신러닝 분류 알고리즘입니다.
* **K-Means Clustering (K-평균 군집화):** 수많은 데이터를 그룹으로 묶을 때, 데이터 점들과 군집의 중심점(Centroid) 사이의 유클리드 거리를 측정하여 가장 가까운 중심점의 그룹으로 편입시킵니다.
* **손실 함수 (Loss Function):** 딥러닝에서 AI가 예측한 값과 실제 정답 사이의 오차를 계산하는 **MSE (Mean Squared Error, 평균 제곱 오차)**는 유클리드 거리 공식에서 루트만 벗겨낸 형태입니다.

#### Python 코드 예시: KNN의 핵심 원리 구현

두 점 사이의 유클리드 거리를 계산하여, 새로운 데이터와 가장 가까운 이웃을 찾는 과정을 파이썬 코드로 구현해 보겠습니다.



```python
import numpy as np

def calculate_euclidean_distance(point1, point2):
    """
    두 다차원 벡터(NumPy 배열) 사이의 유클리드 거리를 계산하는 함수
    """
    # 1. 두 벡터의 차이 계산
    diff = point1 - point2
    # 2. 차이의 제곱합 계산 후 루트 씌우기 (벡터화 연산으로 고속 계산)
    distance = np.sqrt(np.sum(diff ** 2))
    return distance

# 1. 기존 학습 데이터 (예: 과일의 [당도, 크기, 무게])
# 각 데이터가 다차원 공간의 '점'이 됩니다.
dataset = {
    "사과1": np.array([8, 6, 150]),
    "사과2": np.array([9, 5, 140]),
    "배1": np.array([5, 9, 300]),
    "배2": np.array([6, 8, 280])
}

# 2. 분류해야 할 정체불명의 새로운 과일 데이터 등장
new_fruit = np.array([8.5, 5.5, 145])

print(f"새로운 과일의 속성: {new_fruit}\n")

# 3. 새로운 과일과 기존 데이터들 사이의 유클리드 거리 측정
distances = {}
for name, data_point in dataset.items():
    dist = calculate_euclidean_distance(new_fruit, data_point)
    distances[name] = dist
    print(f"'{name}'(와의) 유클리드 거리: {dist:.2f}")

# 4. 가장 거리가 짧은(가까운) 데이터 찾기
nearest_neighbor = min(distances, key=distances.get)
print(f"\n결론: 유클리드 거리가 가장 짧은 데이터는 '{nearest_neighbor}'입니다.")
print("따라서 AI는 이 새로운 과일을 '사과'로 분류합니다.")
```


기원전 300년의 고대 그리스 수학자가 나일강의 토지를 재기 위해 그려낸 단순한 직각삼각형의 원리가, 수천 년이 지난 지금 다차원 인공지능 공간에서 과일을 분류하고 사용자의 취향을 분석하는 알고리즘의 심장으로 뛰고 있습니다. 수학의 본질은 변하지 않으며, 차원만 확장될 뿐입니다.
