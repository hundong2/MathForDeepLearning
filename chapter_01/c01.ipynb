# 확률

## 표본 공간 (sample space)

- 주어진 한 사건의 모든 가능한 결과(OUTCOME)을 나타내는 이산 집합 ( decrete set ) 또는, 연속 구간( continuous range ) 이다. 
- 가능한 모든 결과를 대표
- 개별 사건은 표본 공간의 한 부분 집합, 즉, 표본 공간에서 뽑은 하나의 표본(sample)에 해당. 

### Example

- 동전 앞(head), 뒤(tail)의 경우 표본 공간 `{ H, T }`
- 주사위 : `{1,2,3,4,5,6}`

## 사건 ( event )

- 무언가가 일어나는 것 

## Likelihood 가능도

- 어떠한 사건이 일어날 가능성이 어느 정도인지를 나타내는 값



## 확률 변수 

- 동전 하나 던지기의 결과를 X라는 변수라고 표기할 때, 이 `X`를 `확률변수(random variable)` 이라고 부른다. 
- 표본 공간이 이산적일 때, X는 `이산 확률 변수(discrete random variable)`이라고 부른다. 
- $ P(X=앞면) = P(X = 뒷면 ) = 0.5 $
- 연속 표본 공간에서 값을 취하는 확률 변수 X는 `연속 확률 변수(continuous random variable)`이라 부르고, `x`처럼 소문자로 표기한다.  

## 확률의 법칙 

### 단일 사건의 확률 

- 표본 공간 임의의 사건 A에 대해서 

$$ 0 <= P(A) <= 1 $$

- 표본 공간의 모든 사건 $ A_i $ 에 대해서

$$ \sum_{i} P(A_i) = 1 $$

- 사건 A가 발생할 확률이 P(A)라고 할 때, 그 사건이 발생하지 않을 확률 or 여사건 (complementary event) : $ P(A^c) $

$$ P(\bar{A}) = 1 - P(A) $$

```python
import numpy as np

# 시뮬레이션 횟수 (총 100만 번의 실험)
N = 1000000

# 한 번의 실험에서 동전을 던지는 횟수 (3번)
M = 3

# np.zeros(크기): 주어진 크기만큼 0으로 채워진 배열을 생성합니다.
# 동전을 3번 던졌을 때 앞면이 0회, 1회, 2회, 3회 나올 수 있으므로 크기가 4(M+1)인 배열을 만듭니다.
heads = np.zeros(M+1)
print("heads: %s" % np.array2string(heads))

for i in range(N):
    # np.random.randint(low, high, size): low부터 high 미만의 무작위 정수를 size개 생성합니다.
    # 0과 1 (앞면, 뒷면)을 M개(3개) 무작위로 생성하여 동전 던지기를 시뮬레이션합니다.
    flips = np.random.randint(0, 2, M)
    #print("flips: %s" % np.array2string(flips))
    
    # np.bincount(배열, minlength): 0부터 시작하여 배열 안의 각 정수값이 몇 번 등장했는지 셉니다.
    # 인덱스 0(앞면이라고 가정)의 등장 횟수를 h에, 인덱스 1(뒷면)의 등장 횟수를 _(무시)에 할당합니다.
    h, _ = np.bincount(flips, minlength=2)
    
    # 이번 실험에서 나온 앞면의 개수(h)에 해당하는 인덱스의 값을 1 증가시킵니다.
    heads[h] += 1 

# 전체 발생 횟수를 총 실험 횟수(N)로 나누어 각 횟수별 확률을 계산합니다.
prob = heads / N

# np.array2string(배열): numpy 배열을 출력하기 좋은 문자열 형태로 변환합니다.
print("Probability of getting 0 heads: %s" % np.array2string(prob))
```

# 합의 법칙

- 두 사건 A와 B가 동시에 일어날 수 없을 때, 두 사건을 `상호 배반(mutually exclussive)`
- 만일 두 사건의 발생 확률이 서로 완전히 무관하다면, 두 사건은 서로 `독립(independent)`

$$ P(A \cup B) = P(A) + P(B) - P(A \cap B) $$

# 곱의 법칙
