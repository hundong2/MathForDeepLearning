# 확률

## 표본 공간 (sample space)

- 주어진 한 사건의 모든 가능한 결과(OUTCOME)을 나타내는 이산 집합 ( decrete set ) 또는, 연속 구간( continuous range ) 이다. 
- 가능한 모든 결과를 대표
- 개별 사건은 표본 공간의 한 부분 집합, 즉, 표본 공간에서 뽑은 하나의 표본(sample)에 해당. 

### Example

- 동전 앞(head), 뒤(tail)의 경우 표본 공간 `{ H, T }`
- 주사위 : `{1,2,3,4,5,6}`

## 사건 ( event )

- 무언가가 일어나는 것 

## Likelihood 가능도

- 어떠한 사건이 일어날 가능성이 어느 정도인지를 나타내는 값



## 확률 변수 

- 동전 하나 던지기의 결과를 X라는 변수라고 표기할 때, 이 `X`를 `확률변수(random variable)` 이라고 부른다. 
- 표본 공간이 이산적일 때, X는 `이산 확률 변수(discrete random variable)`이라고 부른다. 
- $ P(X=앞면) = P(X = 뒷면 ) = 0.5 $
- 연속 표본 공간에서 값을 취하는 확률 변수 X는 `연속 확률 변수(continuous random variable)`이라 부르고, `x`처럼 소문자로 표기한다.  

## 확률의 법칙 

### 단일 사건의 확률 

- 표본 공간 임의의 사건 A에 대해서 

$$ 0 <= P(A) <= 1 $$

- 표본 공간의 모든 사건 $ A_i $ 에 대해서

$$ \sum_{i} P(A_i) = 1 $$

- 사건 A가 발생할 확률이 P(A)라고 할 때, 그 사건이 발생하지 않을 확률 or 여사건 (complementary event) : $ P(A^c) $

$$ P(\bar{A}) = 1 - P(A) $$

```python
import numpy as np

# 시뮬레이션 횟수 (총 100만 번의 실험)
N = 1000000

# 한 번의 실험에서 동전을 던지는 횟수 (3번)
M = 3

# np.zeros(크기): 주어진 크기만큼 0으로 채워진 배열을 생성합니다.
# 동전을 3번 던졌을 때 앞면이 0회, 1회, 2회, 3회 나올 수 있으므로 크기가 4(M+1)인 배열을 만듭니다.
heads = np.zeros(M+1)
print("heads: %s" % np.array2string(heads))

for i in range(N):
    # np.random.randint(low, high, size): low부터 high 미만의 무작위 정수를 size개 생성합니다.
    # 0과 1 (앞면, 뒷면)을 M개(3개) 무작위로 생성하여 동전 던지기를 시뮬레이션합니다.
    flips = np.random.randint(0, 2, M)
    #print("flips: %s" % np.array2string(flips))
    
    # np.bincount(배열, minlength): 0부터 시작하여 배열 안의 각 정수값이 몇 번 등장했는지 셉니다.
    # 인덱스 0(앞면이라고 가정)의 등장 횟수를 h에, 인덱스 1(뒷면)의 등장 횟수를 _(무시)에 할당합니다.
    h, _ = np.bincount(flips, minlength=2)
    
    # 이번 실험에서 나온 앞면의 개수(h)에 해당하는 인덱스의 값을 1 증가시킵니다.
    heads[h] += 1 

# 전체 발생 횟수를 총 실험 횟수(N)로 나누어 각 횟수별 확률을 계산합니다.
prob = heads / N

# np.array2string(배열): numpy 배열을 출력하기 좋은 문자열 형태로 변환합니다.
print("Probability of getting 0 heads: %s" % np.array2string(prob))
```

# 합의 법칙

- 두 사건 A와 B가 동시에 일어날 수 없을 때, 두 사건을 `상호 배반(mutually exclussive)`
- 만일 두 사건의 발생 확률이 서로 완전히 무관하다면, 두 사건은 서로 `독립(independent)`

$$ P(A \cup B) = P(A) + P(B) - P(A \cap B) $$

```python
#무작위로 뽑은 두사람의 생일이 같을 확률
match=0
for i in range(100000):
    a = np.random.randint(0, 364)
    b = np.random.randint(0, 364)
    if( a == b):
        match += 1
print("Probability of two people sharing a birthday: %s" % (match/100000))
```

```python
# 생일이 같을 확률이 50%를 넘기 위한 사람 수 
for m in range(2, 30):
    matches=0
    for n in range(10000):
        match = 0
        b = np.random.randint(0, 364, m)
        for i in range(m):
            for j in range(i + 1, m):
                if b[i] == b[j]:
                    match += 1                
        if match != 0:
            matches += 1
print("%d2 %0.6f" % (m, matches/10000))
```

# 조건부 확률 

$$ P(A \mid B) $$

$$ P(A \mid B) = \frac{P(A \cap B)}{P(B)} $$

- B가 일어 났을때, A가 일어날 확률 

- [total probability](./total_probability.md)  


결합 확률(Joint Probability)과 주변 확률(Marginal Probability)은 두 개 이상의 사건(또는 확률 변수)이 얽혀 있을 때 사용하는 개념입니다. 

이 두 가지는 **'엑셀 표(Table)'를 상상**하면 아주 직관적으로 이해할 수 있습니다. 

예를 들어, 전체 학생 100명을 **'성별(남/녀)'**과 **'안경 착용 여부(O/X)'**로 나누어 표를 만들었다고 가정해 보겠습니다.

| 구분 | 안경 씀 (O) | 안경 안 씀 (X) | **합계 (가장자리)** |
| :--- | :---: | :---: | :---: |
| **남자** | 20명 | 30명 | **50명** |
| **여자** | 10명 | 40명 | **50명** |
| **합계** | **30명** | **70명** | **100명 (전체)** |

---

### 1. 결합 확률 (Joint Probability) : "교차점"
*   **의미:** 두 사건 A와 B가 **동시에** 일어날 확률을 말합니다. (사건들의 교집합)
*   **직관적 해석:** 위 표에서 **'안쪽의 특정 칸'** 하나를 콕 집어내는 것입니다.
*   **예시:** "한 명을 뽑았을 때, **남자이면서 동시에 안경을 쓴** 학생일 확률은?"
    *   $P(A, B)$ 또는 $P(A \cap B)$ 로 표기합니다.
    *   위 사례에서는 100명 중 20명이므로 $20/100 = 0.2$ 가 됩니다.

### 2. 주변 확률 (Marginal Probability) : "가장자리의 합계"
*   **의미:** 다른 사건이 어떻게 되든 전혀 상관하지 않고, **오직 내가 관심 있는 특정 사건 하나**만 일어날 확률을 말합니다.
*   **직관적 해석:** 위 표의 **'가장자리(Margin)에 적힌 합계'**를 의미합니다. 표의 여백(가장자리)에 값을 적어 넣었다고 해서 'Marginal(주변의, 가장자리의)'이라는 이름이 붙었습니다.
*   **예시:** 안경을 썼든 안 썼든 상관없이, 한 명을 뽑았을 때 그냥 **남자일 확률**은?"
    *   그냥 $P(A)$ 라고 표기합니다.
    *   가장자리 합계인 50명을 전체 100명으로 나눈 $50/100 = 0.5$ 가 됩니다.

---

### 3. 두 확률의 관계 (수식 연결)
위 표를 자세히 보면 재미있는 사실을 알 수 있습니다.
"남자인 확률(주변 확률)"은 "남자이면서 안경 쓴 확률(결합 확률)"과 "남자이면서 안경 안 쓴 확률(결합 확률)"을 **더한 것**과 같습니다.

이를 수학 기호로 멋지게 쓰면 아래와 같습니다.

> **$$ P(A) = \sum_{B} P(A, B) $$**
> (해석: A가 일어날 주변 확률은, B의 모든 경우에 대한 결합 확률들을 전부 더한 것과 같다.)

우리는 앞에서 조건부 확률의 성질인 $P(A, B) = P(A \mid B)P(B)$를 배웠습니다. 이걸 위 식에 쏙 넣어보면 어떨까요?

> **$$ P(A) = \sum_{B} P(A \mid B)P(B) $$**

짜잔! **이것이 바로 우리가 방금 전에 배웠던 '전체 확률의 법칙'입니다.** 
즉, 주변 확률을 구하는 과정 자체가 결국 전체 확률의 법칙을 적용하는 것과 완전히 동일한 이야기입니다.

**확률의 연쇄 법칙 (Chain Rule of Probability)**은 2개 이상의 여러 사건이 동시에 또는 연달아 일어날 때의 **결합 확률(Joint Probability)을 조건부 확률들의 곱으로 쪼개어 계산**할 수 있게 해주는 법칙입니다.

앞서 다룬 조건부 확률의 정의 $P(A \mid B) = \frac{P(A, B)}{P(B)}$ 를 살짝 이항하면 $P(A, B) = P(B)P(A \mid B)$ 가 되는데, 이것이 연쇄 법칙의 가장 기본 형태입니다.

### 1. 직관적 이해 (2~3개 사건의 경우)

여러 사건이 순차적으로 일어난다고 상상해 보세요. 
예를 들어, **"비가 오고(A), 우산을 챙기고(B), 회사에 지각할(C)"** 결합 확률 $P(A, B, C)$를 구하고 싶습니다. 한 번에 구하기엔 너무 복잡하죠. 이렇게 이야기로 풀어서 곱하면 됩니다.

1. 먼저 **비가 올 확률**을 구합니다. $\rightarrow P(A)$
2. 비가 왔다는 전제하에, **우산을 챙길 확률**을 곱합니다. $\rightarrow P(B \mid A)$
3. 비가 왔고 우산도 챙겼다는 전제하에, **지각할 확률**을 곱합니다. $\rightarrow P(C \mid A, B)$

이것을 수식으로 나타내면 다음과 같이 사슬(Chain)처럼 꼬리를 물고 이어집니다.

> **$$ P(A, B, C) = P(A) P(B \mid A) P(C \mid A, B) $$**

---

### 2. 일반화 (Generalization)

사건이 $A_1, A_2, A_3, ..., A_n$ 처럼 $n$개로 무한히 길어지더라도 이 법칙은 똑같이 적용할 수 있습니다. 
"첫 번째 사건이 일어나고, 그 다음엔 앞의 사건들을 조건으로 깔고 다음 사건이 일어나고..." 를 반복하는 것이죠.

수식으로 전개하면 아래와 같습니다.


$$ P(A_1, A_2, \dots, A_n) = P(A_1) P(A_2 \mid A_1) P(A_3 \mid A_1, A_2) \dots P(A_n \mid A_1, \dots, A_{n-1}) $$


수학에서는 이렇게 여러 개를 연속해서 곱하는 것을 나타낼 때 덧셈의 $\sum$(시그마) 대신 **곱셈의 기호인 $\prod$(파이)** 를 사용합니다. 이를 이용해 아주 간결하게 일반화된 수식으로 표현하면 아래와 같습니다.


$$ P(A_1, \dots, A_n) = \prod_{i=1}^{n} P(A_i \mid A_1, \dots, A_{i-1}) $$

*(단, $i=1$일 때는 조건부가 없는 $P(A_1)$을 의미합니다.)*
